### For monthly

```python
primary_etccdi_data = pd.read_excel('G:/fresh_start/main_thesis/data/saved_data/etccdi_climate_indices.xlsx')
#etccdi_season_result_df
```

```python
import pandas as pd
import pymannkendall as mk

# Assuming you have your data in a DataFrame called primary_etccdi_data

# Create an empty list to store the results as dictionaries
results_list = []

# Get unique Geo_regions
unique_regions = primary_etccdi_data['Geo_region'].unique()
unique_months = primary_etccdi_data['Month'].unique()
# Get the temperature indices you want to test
temperature_indices = ['TNn', 'TNx', 'TXn', 'TXx', 'TN10p', 'TN90p', 'TX10p', 'TX90p']

# Loop through each Geo_region and temperature index
for month in unique_months:
  for region in unique_regions:
      for index in temperature_indices:
          # Filter data for the current region and index
          
          filtered_data = primary_etccdi_data[(primary_etccdi_data['Geo_region'] == region) & (primary_etccdi_data['Month'] == month)]

          # Perform the Mann-Kendall test
          result = mk.original_test(filtered_data[index], alpha=0.05)
          
          # Append the results to the list as a dictionary
          results_list.append({
              'Geo_region': region,
              'Month' : month,
              'Index': index,
              'Slope': result.slope,
              'Intercept': result.intercept,
              'Trend': result.trend,
              'P-Value': result.p,
              'Z-Value': result.z
          })

# Convert the list of dictionaries into a DataFrame
results_df = pd.DataFrame(results_list)

# Now, results_df contains the test results for each Geo_region and temperature index
print(results_df)
```

```python
results_df.to_csv('G:/fresh_start/main_thesis/data/saved_data/etccdi_monthly_trend_final.csv')
```
