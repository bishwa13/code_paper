```python
import pandas as pd
import numpy as np
import pymannkendall as mk
from scipy.stats import spearmanr
```

```python
data_import = pd.read_csv(r"G:\fresh_start\paper\code_paper\main_data\final_data\station_annual_mean_data.csv")
```

```python

# for annual

# Initialize a dictionary to store Mann-Kendall results
mann_kendall_results = {}

# Loop through each station
for station in data_import['station'].unique():
    # Filter data for the current station
    df_filtered = data_import[data_import['station'] == station]
    
    # Define columns to test
    selected_columns = ['Tmax', 'Tmin', 'Tavg']
    
    # Perform the Mann-Kendall Trend Test for each column
    for column in selected_columns:
        # Extract the data for the column
        data = df_filtered[column].values
        
        # Perform the Mann-Kendall Trend Test
        result = mk.original_test(data)
        
        # Store the result in the dictionary with a unique key
        key = (station, column)
        mann_kendall_results[key] = {
            'Slope': result.slope,
            'Intercept': result.intercept,
            'Trend': result.trend,
            'P-Value': result.p,
            'Z-Value': result.z,
        }

# Convert results to a DataFrame
results_df = pd.DataFrame(mann_kendall_results).T.reset_index()
results_df.columns = ['Station', 'Column', 'Slope', 'Intercept', 'Trend', 'P-Value', 'Z-Value']

# Print results
print(results_df)
```
