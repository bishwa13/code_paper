```python
import pandas as pd
import openpyxl 
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
filtered_df_1992 = pd.read_csv(r'G:\fresh_start\paper\code_paper\main_data\raw_data\filtered_df_1992_filled_lapse.csv')
```

```python
def plot_station_data(df):
    stations = df['station'].unique()
    
    for station in stations:
        station_data = df[df['station'] == station]
        
        plt.figure(figsize=(10, 6))
        plt.plot(station_data['date'], station_data['Tmin'], label='Tmin', marker='o')
        plt.plot(station_data['date'], station_data['Tmax'], label='Tmax', marker='x')
        
        plt.title(f'Station {station} Temperature Trends')
        plt.xlabel('Date')
        plt.ylabel('Temperature (Â°C)')
        plt.legend()
        plt.grid(True)
        
        plt.show()

```

```python
# Call the function to plot data for each station
plot_station_data(filtered_df_1992)
```
