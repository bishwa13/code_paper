```python


initial_1962_average_dict = {}

# Iterate through the keys in both dictionaries
for key in initial_max_daily_1962_temp_dict.keys():
    if key in initial_min_daily_1962_temp_dict:
        
        df_max = initial_max_daily_1962_temp_dict[key]
        #df_max1 =df_max.reset_index()
        df_min = initial_min_daily_1962_temp_dict[key]
        #df_min1 = df_min.reset_index()
        # Calculate the average of the DataFrames with the same key
        averaged_df = (df_max + df_min ) / 2
        
        # Ensure the averaged DataFrame has the same column names and index as dict1
        #averaged_df.columns = initial_max_daily_2007_temp_dict[key].columns
        #averaged_df.index = initial_max_daily_2007_temp_dict[key].index
        
        #print(df_max)
        averaged_df = averaged_df.set_index('Station')
        
        df_melted_df = averaged_df.reset_index().melt(id_vars =['Station'], 
                                    var_name ='Date', value_name = 'avg T')
    
        df_melted_df = df_melted_df[['Date', 'Station','avg T']]
    
        initial_1962_average_dict[key] = df_melted_df




initial_1987_average_dict = {}

# Iterate through the keys in both dictionaries
for key in initial_max_daily_1987_temp_dict.keys():
    if key in initial_min_daily_1987_temp_dict:
        
        df_max = initial_max_daily_1987_temp_dict[key]
        #df_max1 =df_max.reset_index()
        df_min = initial_min_daily_1987_temp_dict[key]
        #df_min1 = df_min.reset_index()
        # Calculate the average of the DataFrames with the same key
        averaged_df = (df_max + df_min ) / 2
        
        # Ensure the averaged DataFrame has the same column names and index as dict1
        #averaged_df.columns = initial_max_daily_2007_temp_dict[key].columns
        #averaged_df.index = initial_max_daily_2007_temp_dict[key].index
        
        #print(df_max)
        averaged_df = averaged_df.set_index('Station')
        
        df_melted_df = averaged_df.reset_index().melt(id_vars =['Station'], 
                                    var_name ='Date', value_name = 'avg T')
    
        df_melted_df = df_melted_df[['Date', 'Station','avg T']]
    
        initial_1987_average_dict[key] = df_melted_df




initial_2007_average_dict = {}

# Iterate through the keys in both dictionaries
for key in initial_max_daily_2007_temp_dict.keys():
    if key in initial_min_daily_2007_temp_dict:
        
        df_max = initial_max_daily_2007_temp_dict[key]
        #df_max1 =df_max.reset_index()
        df_min = initial_min_daily_2007_temp_dict[key]
        #df_min1 = df_min.reset_index()
        # Calculate the average of the DataFrames with the same key
        averaged_df = (df_max + df_min ) / 2
        
        # Ensure the averaged DataFrame has the same column names and index as dict1
        #averaged_df.columns = initial_max_daily_2007_temp_dict[key].columns
        #averaged_df.index = initial_max_daily_2007_temp_dict[key].index
        
        #print(df_max)
        averaged_df = averaged_df.set_index('Station')
        
        df_melted_df = averaged_df.reset_index().melt(id_vars =['Station'], 
                                    var_name ='Date', value_name = 'avg T')
    
        df_melted_df = df_melted_df[['Date', 'Station','avg T']]
    
        initial_2007_average_dict[key] = df_melted_df




```

#### For seasonal average

```python
seasons_df = ['Winter','Premonsoon','Monsoon', 'Postmonsoon' ]
#creating functions to extract seasonal data
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Premonsoon'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    elif month in [10, 11]:
        return 'Postmonsoon'
    else:
        return 'Unknown'
```

```python

seasonal_df_1962 = concatenated_df_filled_1962
seasonal_df_1962['Season'] = seasonal_df_1962['Month'].apply(get_season)


seasonal_df_1987 = concatenated_df_filled_1987

seasonal_df_1987['Season'] = seasonal_df_1987['Month'].apply(get_season)


seasonal_df_2007 = concatenated_df_filled_2007

seasonal_df_2007['Season'] = seasonal_df_2007['Month'].apply(get_season)
```

```python


# Group by 'Station', month, and year, and calculate the maximum 'Max T'
grouped_1962_seasonal = seasonal_df_1962.groupby(['Station', 'Season', 'Year'])['avg T'].mean().reset_index()

# Create a dictionary to store DataFrames for each month
filled_1962_seasonal_avg_dict = {}
for season in seasons_df:
    # Filter data for the current month
    seasonal_data = grouped_1962_seasonal[grouped_1962_seasonal['Season'] == season]
    
    
    seasonal_avg_temp = seasonal_data.pivot(index = 'Station', columns=['Year'], values='avg T')
    
    filled_1962_seasonal_avg_dict[season] = seasonal_avg_temp



# Group by 'Station', month, and year, and calculate the maximum 'Max T'
grouped_1987_seasonal = seasonal_df_1987.groupby(['Station', 'Season', 'Year'])['avg T'].mean().reset_index()

# Create a dictionary to store DataFrames for each month
filled_1987_seasonal_avg_dict = {}
for season in seasons_df:
    # Filter data for the current month
    seasonal_data = grouped_1987_seasonal[grouped_1987_seasonal['Season'] == season]
    
    
    seasonal_avg_temp = seasonal_data.pivot(index = 'Station', columns=['Year'], values='avg T')
    
    filled_1987_seasonal_avg_dict[season] = seasonal_avg_temp



# Group by 'Station', month, and year, and calculate the maximum 'Max T'
grouped_2007_seasonal = seasonal_df_2007.groupby(['Station', 'Season', 'Year'])['avg T'].mean().reset_index()

# Create a dictionary to store DataFrames for each month
filled_2007_seasonal_avg_dict = {}
for season in seasons_df:
    # Filter data for the current month
    seasonal_data = grouped_2007_seasonal[grouped_2007_seasonal['Season'] == season]
    
    
    seasonal_avg_temp = seasonal_data.pivot(index = 'Station', columns=['Year'], values='avg T')
    
    filled_2007_seasonal_avg_dict[season] = seasonal_avg_temp
```
