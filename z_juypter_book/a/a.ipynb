```python
import numpy as np
import matplotlib.pyplot as plt

# Generate made-up data
dates = np.array(['1/1/1972', '1/1/1973', '1/1/1974', '1/1/1975', '1/1/1976'])
time_points = np.arange(len(dates))

# Made-up discharge data (m³/s)
observed_discharge = np.array([50, 120, 80, 150, 90])  # Observed
simulated_discharge = np.array([60, 110, 90, 140, 100])  # Simulated

# Made-up precipitation data (mm)
precipitation = np.array([0, 200, 100, 300, 150])

# Create figure and axis objects with a secondary y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot discharge data on the first y-axis
ax1.plot(time_points, observed_discharge, 'o-', label='Q_Observed', linestyle='--', color='blue')
ax1.plot(time_points, simulated_discharge, 'o-', label='Q_Simulated', color='orange')
ax1.set_xlabel('Date')
ax1.set_ylabel('Discharge (m³/s)', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xticks(time_points)
ax1.set_xticklabels(dates, rotation=45)
ax1.grid(True)

# Add calibration and validation annotation
ax1.axvline(x=2, color='gray', linestyle='--', linewidth=1)
ax1.text(1, 200, 'Calibration', rotation=90, va='center')
ax1.text(3, 200, 'Validation', rotation=90, va='center')

# Create a second y-axis for precipitation (inverted)
ax2 = ax1.twinx()
# Invert the y-axis for precipitation
ax2.invert_yaxis()
# Plot bars from 0 downward by making precipitation values negative
ax2.bar(time_points, -precipitation, alpha=0.3, color='green', label='Precipitation', width=0.4)
ax2.set_ylabel('Precipitation (mm)', color='green')
# Make y-tick labels positive even though we plot negative values
ax2.set_yticklabels([abs(int(x)) for x in ax2.get_yticks()])
ax2.tick_params(axis='y', labelcolor='green')

# Add title and legend
plt.title('Comparison of Observed and Simulated Hydrograph during Calibration and Validation for 1972-1976 of the KRB\n(A. Shrestha et al. / Journal of Cleaner Production 279 (2021) 123483)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
